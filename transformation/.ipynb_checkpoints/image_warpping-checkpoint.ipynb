# Image Warpping

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
IMAGE_DIR = "./images/"
image = 'grid.jpg'
```

```python
img = cv2.imread(IMAGE_DIR+image)
plt.imshow(img)
width, height, ch = img.shape
```

```python
cv2.circle(img, (447, 90), 10, (0, 0, 255), -1)
cv2.circle(img, (83, 90), 10, (0, 0, 255), -1)
cv2.circle(img, (83, 472), 10, (0, 0, 255), -1)

plt.imshow(img)
```

```python
ptsl = np.float32([[83, 90], [447,90], [83, 472]])
pts2 = np.float32([[30, 90], [447,90], [83, 472]])
```

```python
ptsl = np.float32([[83, 90], [447,90], [83, 472]])
pts2 = np.float32([[30, 90], [447,90], [83, 472]])
affined_matrix = cv2.getAffineTransform(ptsl, pts2)
affined_img = cv2.warpAffine(img, affined_matrix, (height, width))
```

```python
fig = plt.figure(figsize=[20, 10])
ax1 = fig.add_subplot(121)
ax1.imshow(img)
ax1 = fig.add_subplot(122)
ax1.imshow(affined_img)
```
